# Bar Graphs

To create a bar graph, we are going to import data from the 2024-2025 NBA season.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NBA/player/nba_player_stats_2024-2025.csv')
df
```

The code below will filter the data to compare two players with similar scoring averages but different playing styles.

```python
player1 = "Pascal Siakam"
player2 = "Stephen Curry"
comparison = df[(df['Name'] == player1) | (df['Name'] == player2)]
comparison
```

Now that we have filtered comparison data, we can create a bar graph.

```python
import plotly.express as px
px.bar(comparison, x='Name', y=['FG%', '3P%', 'PTS', 'AST', 'REB'], barmode='group', title=f'Comparing {player2} to {player1}')
```
