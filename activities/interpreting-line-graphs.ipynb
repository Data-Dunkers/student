# Interpreting Line Graphs

To create a [line graph](https://en.wikipedia.org/wiki/Line_chart), we are going to import data about Pascal Siakam's NBA career.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NBA/Pascal_Siakam.csv')
df
```

 The code below will create a line graph that shows how many games Pascal Siakam played in each season from 2015-2016 to 2024-2025.
 
 Line charts are often used to show trends over time, so we put time (season) on the x-axis.

```python
import plotly.express as px
px.line(df, x="SEASON_ID", y="GP", title="Games Played per Season", labels={"SEASON_ID": "Season", "GP": "Games Played"})
```

Hover over different points to see their values.
