# Interpreting Scatterplots

To create a [scatterplot](https://en.wikipedia.org/wiki/Scatter_plot), we are going to import data from the 2024-2025 NBA season.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NBA/player/nba_player_stats_2024-2025.csv')
df
```

 The code below will create a scatterplot that shows how NBA players' field goal percentages relate to the points they scored.

```python
import plotly.express as px
px.scatter(df, x="PTS", y="FG%", hover_name="Name", title="Field Goal Percentage vs Points", labels={"PTS": "Points per Game", "FG%": "Field Goal Percentage"})
```

Identify some players that stand out from the rest. Hover over different points to see player names and statistics.
