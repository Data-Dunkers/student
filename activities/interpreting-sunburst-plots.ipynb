# Interpreting Sunburst Charts

A [sunburst chart](https://plotly.com/python/sunburst-charts/) is like a pie chart, but it can show multiple levels of data.

We are going to import data from the 2024-2025 Indiana Pacers NBA season.

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NBA/team/2024-2025/IND_2024-2025_players.csv")
df
```

The code below will create a sunburst chart to visualize the distribution of points and shots.

```python
import plotly.express as px

df = df[df['Name'] != 'Total']
df = df.sort_values(by="PTS", ascending=False)
df_melted = df.melt(id_vars=['Name'], value_vars=['2PM', '3PM', 'FTM'], var_name='ShotType', value_name='Count')

px.sunburst(df_melted, path=['Name', 'ShotType'], values='Count', title='Sunburst of Shot Distribution by Player')
```

The data can also be visualized using a [treemap](https://plotly.com/python/treemaps/).

```python
px.treemap(df_melted, path=['Name', 'ShotType'], values='Count', title='Treemap of Shot Distribution by Player')
```
