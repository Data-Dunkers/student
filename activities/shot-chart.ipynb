# Shot Chart

Basketball shot charts show where players take their shots on the court and how successful they were. This visualization helps you understand players' shooting patterns and effectiveness.

We'll start by importing and displaying the shot data for the [Indiana Pacers](https://en.wikipedia.org/wiki/Indiana_Pacers) from the 2024-2025 season.

```python
import pandas as pd
shots = pd.read_csv('https://raw.githubusercontent.com/Data-Dunkers/data/refs/heads/main/NBA/pacers-shots-2024-2025.csv')
shots
```

Now that we have the data, we can create a visualization of the court locations (x and ) of all those shots.

```python
import plotly.express as px
px.scatter(shots, x='LOC_X', y='LOC_Y', height=1000, width=800, color='PLAYER_NAME')
```

We can also add some more data to the visualization from other columns in the dataframe.

```python
px.scatter(shots, x='LOC_X', y='LOC_Y', height=1100, width=900, color='ACTION_TYPE', hover_data=['PLAYER_NAME', 'SHOT_MADE_FLAG', 'SHOT_DISTANCE', 'PERIOD'])
```

You can hover over points to see player names and other information. Click or double-click on the shot types in the legend to filter the data.
