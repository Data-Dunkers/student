## Data Analysis

Run the following code cell to import and display the whole data set.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

url = 'https://docs.google.com/spreadsheets/d/1WQkl5Aj6PKoXHLGxaX-XBPe_Ds5jRFExPHTavbwrKV0/export?format=csv'
df = pd.read_csv(url)
df.head()
```

Now that you have the data, run the following code cell to calculate the average (`mean`) of `Shots made` at each distance.

```python
distance = df.columns[2]
averages = df.groupby(distance)['Shots made'].mean()

averages
```

Run the following code cell to make a bar chart from those calculated averages.

```python
px.bar(averages, y='Shots made', title='Average Shots Made at Each Distance')
```

```python
me = 'RC78'
my_data = df[df['Initials'] == me]
fig = px.bar(my_data, x='Shot distance (feet)', y='Shots made', title='Shots Made at Each Distance', barmode='group')
fig.add_trace(go.Bar(x=averages.index, y=averages.values, name='Average'))
fig.show()
```
